# Manipulações de dados com Pandas II
## Desafio Complementar

1. Usando o arquivo "best-selling game consoles.xlsx" conte quantos consoles tem na base por companhias.

```python
import pandas as pd
```

```python
consoles = pd.read_excel("best-selling game consoles.xlsx",
                      sheet_name="consoles")
consoles.head()
```

```python
companhias = consoles.groupby('Company')[['Company']].count()
companhias
```

2. Ordene o resultado anterior pela quantidade de consoles por companhia de forma decrescente. 

```python
companhias.columns = ['qtd']
companhias = (companhias
              .sort_values('qtd',ascending=False)
              .reset_index())
companhias
```

3. Sobre o dataframe anterior, adicione a coluna "Market Value" presente no tabela do arquivo "company.txt".

```python
market_value = pd.read_csv("company.txt")
market_value
```

```python
companhias_mv = companhias.merge(market_value,on='Company',how='left')
companhias_mv
```

4. Qual o valor médio de Market Value por console?

```python
soma = companhias_mv.sum()
soma
```

```python
print("Valor Médio: ", soma['Market Value']/soma['qtd'])
```
