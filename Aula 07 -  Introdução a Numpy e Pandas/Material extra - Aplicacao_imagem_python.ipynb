```python
### brincando com imagem
from skimage import io
photo = io.imread('python.jpg')
type(photo)
```

```python
### dimensão da imagem
photo.ndim
```

```python
### tamanho de cada dimensão
photo.shape
```

```python
### plotando a imagem
import matplotlib.pyplot as plt
plt.imshow(photo)
```

```python
### para cada elemento da "matriz" teríamos três valores que corresponderiam a cada um dos espectros RGB
photo[0,0,:]
```

```python
photo[:,:,0]
```

```python
plt.imshow(photo[::-1])
```

```python
plt.imshow(photo[:, ::-1])
```

```python
plt.imshow(photo[0:192, 100:320, ::-1])
```
